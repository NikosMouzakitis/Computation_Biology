```python
# function used to find the most 
# frequent k-mers in
# a nucleotides series

def PatternCount(Pattern, Text):
    count = 0
    for i in range(len(Text) - len(Pattern)+1):
       
        if(Text[i:i+len(Pattern)] == Pattern):
            count = count + 1
    return count
```

```python
sample="CGATATATATCCATAG"
pattern="ATA"
print(PatternCount(pattern, sample))
```

```python
# create frequency map for all
# the k-mers into a given string

def FrequencyMap(Text, k):
    freq={} #empty dictionary
    n=len(Text)
    
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        freq[Pattern]=0
    
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        freq[Pattern]+=1
    
    
    return freq
    
```

```python
k = 3
print(FrequencyMap("CGATATATCCATAG",k))
```

```python

words = []
freq = FrequencyMap("CGATATATCCATAG",k)

#max value from the freq dictionary.
m = max(freq.values())
for keys in freq:
    if(freq[keys] == m):
        words.append(keys)
        
print(words)
```

```python
def FrequentWords(Text, k):
    words=[]
    freq=FrequencyMap(Text,k)
    m=max(freq.values())
    
    for keys in freq:
        if(freq[keys]==m):
            words.append(keys)
    return words
```

```python
def Complement(Pattern):
    s=""
    for i in range(0, len(Pattern)):
        if(Pattern[i] == "A"):
            cmp = "T"
        elif(Pattern[i] == "T"):
            cmp = "A"
        elif(Pattern[i] == "G"):
            cmp = "C"
        elif(Pattern[i] == "C"):
            cmp = "G"
        s=s+cmp
        
    return s            
```

```python
print(Complement("AAAACCCGGT"))
```

```python
def ReverseComplement(Pattern):
    Pattern=Reverse(Pattern)
    Pattern=Complement(Pattern)
    return Pattern
```

```python
def SkewArray(Genome):
    array = []
    Skew = 0
    array.append(0)
    
    for i in Genome:
        if ( (i == "A") or (i == "T") ):
            array.append(Skew)
        elif ( i == "G"):
            Skew+=1
            array.append(Skew)
        elif( i == "C"):
            Skew-=1
            array.append(Skew)
    return array
```

```python
p = SkewArray("CATGGGCATCGGCCATACGCC")
print(p)
```

```python
def MinimumSkew(Genome):
    array = SkewArray(Genome)
    positions = []
    minarray=min(array)
    count = 0 # to get the index in array.
    
    for i in array:
        if(i==minarray):
            positions.append(count)
        count+=1
    return positions
```

```python
p = MinimumSkew("TAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT")
for i in p:
    print(i)
```

```python
# calculation of hamming distance 
# between two genomes
def HammingDistance(g1,g2):
    d = 0
    for i in range(0,len(g1)):
        if(g1[i] != g2[i]):
            d+=1
    return d
```
